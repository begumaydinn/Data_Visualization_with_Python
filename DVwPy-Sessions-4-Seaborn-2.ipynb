```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
print(sns.get_dataset_names())
```

```python
tips= sns.load_dataset("tips")
tips.head()
```

```python
sns.countplot(x="day", data = tips);
```

```python
fig, ax_tip = plt.subplots()
ax_tip = sns.countplot(x="day", data = tips);
# sns.countplot(x="day", data = tips, ax =ax_tip );
for p in ax_tip.patches:
    ax_tip.annotate((p.get_height()), (p.get_x()+0.3, p.get_height()+1));
```

```python
sns.countplot(tips["day"]);
```

```python
fig, ax_tip = plt.subplots()
ax_tip = sns.countplot(x="day", data = tips, hue="smoker");
# sns.countplot(x="day", data = tips, ax =ax_tip );
for p in ax_tip.patches:
    ax_tip.annotate((p.get_height()), (p.get_x()+0.1, p.get_height()+1));
```

```python
fig, ax_tip = plt.subplots()
ax_tip = sns.countplot(x="day", data = tips, hue="size");
# sns.countplot(x="day", data = tips, ax =ax_tip );
for p in ax_tip.patches:
    ax_tip.annotate((p.get_height()), (p.get_x()+0.1, p.get_height()+1));
```

```python
fig, ax_tip = plt.subplots()
ax_tip = sns.countplot(x="day", data = tips);
# sns.countplot(x="day", data = tips, ax = ax_tip );
for p in ax_tip.patches:
    ax_tip.annotate((p.get_height()), (p.get_x()+0.3, p.get_height()+1));
    ax_tip.annotate("%"+str(round((p.get_height()/tips.day.count())*100,2)), (p.get_x()+0.1, p.get_height()-5));
```

## Swarmplot

```python
plt.figure(figsize=(12,6))
sns.swarmplot("total_bill", data = tips)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.swarmplot("total_bill", data = tips, size=10)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.swarmplot("total_bill", data = tips, size=10, y="smoker", hue="sex")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.swarmplot("total_bill", data = tips, size=10, y="smoker", hue="sex", dodge=True)
plt.show()
```
