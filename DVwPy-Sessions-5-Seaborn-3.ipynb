```python
sns.PairGrid(penguins)
```

```python
g = sns.PairGrid(penguins)
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(penguins, hue="species")
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(penguins, hue="species")
g.map_diag(sns.kdeplot)
```

```python
g = sns.PairGrid(penguins, hue="species")
g.map_diag(sns.histplot)
```

```python
g = sns.PairGrid(penguins, hue="species")
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot)
g.map_upper(sns.scatterplot)
```

```python
g = sns.PairGrid(penguins)
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot)
g.map_upper(sns.scatterplot)
```

```python
sns.set_palette(["green", "blue", "red"])
g = sns.PairGrid(penguins, hue="species")
g.map_diag(sns.kdeplot)
g.map_lower(sns.kdeplot)
g.map_upper(sns.scatterplot)

g.add_legend()

# saÄŸ alt kÃ¶ÅŸede sÄ±kÄ±ntÄ± var diag_sharey true o yuzden ortak kullanÄ±yor y eksenini
```

```python
sns.set_palette(["green", "blue", "red"])
g = sns.PairGrid(penguins, hue="species", diag_sharey=False)
#y eksenini ortak almasÄ±n diye  eger true olsaydÄ± yukarÄ±daki saÄŸ alt kÃ¶ÅŸe gibi olurdu
g.map_diag(sns.kdeplot)
g.map_lower(sns.kdeplot)
g.map_upper(sns.scatterplot)

g.add_legend()
```
