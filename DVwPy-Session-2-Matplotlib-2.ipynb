```python
tips.day.unique()
```

```python
tips.groupby("day").sum()
```

```python
tips.groupby("day")["total_bill"].sum()
```

```python
day = tips.day.unique() # unique yanlış oldu
total_bill = tips.groupby("day")["total_bill"].sum()
```

```python
fig, ax = plt.subplots()

ax.bar(day, total_bill)
plt.show()  # günler farKlı olduğu için yanlış oldu. buna index eklemek gerek aşağıda yaptık
```

```python
day = tips.groupby("day")["total_bill"].sum().index
tip = tips.groupby("day")["tip"].sum()
fig, ax = plt.subplots()
ax.bar(day,tip)
plt.show()
```

```python
day = tips.groupby("sex")["total_bill"].sum().index
tip = tips.groupby("sex")["tip"].sum()
fig, ax = plt.subplots()
ax.bar(day,tip)
plt.show()
```

```python
sex = tips.groupby("sex")["total_bill"].sum().index
tip = tips.groupby("sex")["total_bill"].sum()
fig, ax = plt.subplots()
ax.bar(sex,tip)
plt.show()
```

```python
day = tips.groupby("day")["total_bill"].sum().index
tip = tips.groupby("day")["tip"].sum()
total = tips.groupby("day")["total_bill"].sum()
fig, ax = plt.subplots()
ax.bar(day, total, label="total" )
ax.bar(day, tip, label="tip" )
plt.legend()
plt.show()
```

```python
plt.bar(tips.groupby('sex').total_bill.sum().index,tips.groupby('sex').total_bill.sum())
```
